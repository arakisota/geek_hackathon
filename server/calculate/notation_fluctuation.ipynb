## どんな方針でやるか
- 先に重複した駅名を一意にする
    - 代表値は最頻値で対応
- 重複しているデータのみを取り出して、変換用のデータフレームを作成する

```python
free_df_dup_grouby = free_df_dup.groupby(["station_name", "lat", "lon"]).count()
free_df_dup_grouby = free_df_dup_grouby.reset_index()
display(free_df_dup_grouby.head())
```

```python
free_df_dup_grouby.query('station_name == "押上〈スカイツリー前〉"')
```

```python
# free_df_dup_groubyのstation_nameで重複している行のうち、countが最も多い行のみを抽出
free_df_dup_groupby_dup = free_df_dup_grouby[free_df_dup_grouby.duplicated(subset='station_name', keep=False)]
free_df_dup_groupby_nodup = free_df_dup_grouby[~free_df_dup_grouby.duplicated(subset='station_name', keep=False)]
free_df_dup_grouby_dup = free_df_dup_groupby_dup.sort_values(by='station_cd', ascending=False)
free_df_dup_grouby_dup = free_df_dup_grouby_dup.drop_duplicates(subset='station_name', keep='first')
free_df_dup_grouby_dup = pd.concat([free_df_dup_grouby_dup, free_df_dup_groupby_nodup])
free_df_dup_grouby_dup.reset_index(drop=True, inplace=True)
display(free_df_dup_grouby_dup.head())
```

```python
free_df_dup_grouby_dup.query('station_name == "押上〈スカイツリー前〉"')[["station_name", "lat", "lon"]]
```

```python
unique_station_lat_lon = free_df_dup_grouby_dup[["station_name", "lat", "lon"]]
unique_station_lat_lon.reset_index(drop=True, inplace=True)
display(unique_station_lat_lon.head())
```

### ここから、駅名ごとでstation_cdを決めたい

```python
station_cd_list = []
for name in unique_station_lat_lon["station_name"]:
    station_cd = free_tokyo_df.query('station_name == @name')["station_cd"].values[0]
    station_cd_list.append(station_cd)
unique_station_lat_lon["station_cd"] = station_cd_list
display(unique_station_lat_lon.head())
```

### ここからは、どのstation_cdが変わったかをまとめたテーブルを作成する

```python
raw_station_cd_list = []
converted_station_cd_list = []
station_name_list = []
for name in unique_station_lat_lon["station_name"]:
    raw_station_cd = free_df_dup.query('station_name == @name')["station_cd"].values
    for cd in raw_station_cd:
        station_name_list.append(name)
        raw_station_cd_list.append(cd)
        converted_station_cd_list.append(unique_station_lat_lon.query("station_name == @name")["station_cd"].values[0])

print(len(raw_station_cd_list))
print(len(converted_station_cd_list))
converted_table = pd.DataFrame({"station_name": station_name_list, "raw_station_cd": raw_station_cd_list, "converted_station_cd": converted_station_cd_list})
display(converted_table.head())
```

```python
for i in range(len(adjacency_result)):
    departure = adjacency_result["departure"][i]
    destination = adjacency_result["destination"][i]
    if departure in bug["departure"].values and destination in bug["destination"].values:
        bug_rows = bug.query("departure == @departure and destination == @destination")
        if not bug_rows.empty:
            adjacency_result.loc[(adjacency_result["departure"] == departure) & (adjacency_result["destination"] == destination), "time"] = bug_rows["time"].values[0]
            adjacency_result.loc[(adjacency_result["departure"] == departure) & (adjacency_result["destination"] == destination), "count"] = 0
            adjacency_result.loc[(adjacency_result["departure"] == departure) & (adjacency_result["destination"] == destination), "fare"] = bug_rows["fare"].values[0]
            if bug_rows["departure"].values[0] == "NaN":
                adjacency_result.loc[(adjacency_result["departure"] == departure) & (adjacency_result["destination"] == destination), "departure"] = bug_rows["departure"].values[0]
            if bug_rows["destination"].values[0] == "NaN":
                adjacency_result.loc[(adjacency_result["departure"] == departure) & (adjacency_result["destination"] == destination), "destination"] = bug_rows["destination"].values[0]

display(adjacency_result.head(30))
```
